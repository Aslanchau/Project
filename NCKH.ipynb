```python
def cronbach_alpha(df_encoded):
    # Số lượng các câu hỏi
    k = df_encoded.shape[1]
    
    # Tính tổng phương sai của các mục
    item_variances = df_encoded.var(axis=0, ddof=1)
    total_variance = df_encoded.sum(axis=1).var(ddof=1)
    
    # Tính Cronbach's Alpha
    alpha = (k / (k - 1)) * (1 - item_variances.sum() / total_variance)
    return alpha

# Tính toán Cronbach's Alpha cho toàn bộ bộ dữ liệu
alpha_total = cronbach_alpha(df_encoded)
print(f"Cronbach's Alpha cho toàn bộ dữ liệu: {alpha_total}")

# Tính toán Cronbach's Alpha khi bỏ từng biến một
alphas = {}
for column in df_encoded.columns:
    df_subset = df_encoded.drop(columns=[column])
    alphas[column] = cronbach_alpha(df_subset)

# In ra danh sách các biến cùng với chỉ số Cronbach's Alpha khi bỏ biến đó
for column, alpha in alphas.items():
    print(f"Cronbach's Alpha khi bỏ '{column}': {alpha}")
```

## ROC AND AUC GRAPH for MODEL COMPARISON

```python
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Assuming 'Creditworthy' is the positive class
label_encoder = LabelEncoder()
test_y_binary = label_encoder.fit_transform(y_test)

fig, ax = plt.subplots(figsize=(8, 6))

models = [
    ("Logistic Regression", LR),
    ("Decision Tree", DT),
    ("Random Forest", RF),
    ("Boosted Tree", XGB),
    ("K-Nearest Neighbor", knn)
]

for name, model in models:
    if name == "Linear Regression":
        y_prob = model.predict(X_test)
    else:
        y_prob = model.predict_proba(X_test)[:, 1]

    fpr, tpr, _ = roc_curve(test_y_binary, y_prob)
    roc_auc = auc(fpr, tpr)

    ax.plot(fpr, tpr, lw=2, label=f'{name} (AUC = {roc_auc:.2f})')

ax.plot([0, 1], [0, 1], color='gray', linestyle='--')
ax.set_xlim([0.0, 1.0])
ax.set_ylim([0.0, 1.05])
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.set_title('Receiver Operating Characteristic (ROC) Curve')
ax.legend(loc='lower right')

plt.show()
```
